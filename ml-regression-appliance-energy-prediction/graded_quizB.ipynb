### QUESTION 12

* From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius(x=T2) and the temperature outside the building (y=T6). What is the R^2 value in two d.p?

```python
#Select the predictor

X = energy_data[['T2']]
X
```

```python
#Select the Target
y = energy_data[['T6']]
y
```

### Splitting our data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
#Importing and declaring our model
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()

#fit the model to the training dataset
lin_model.fit(X_train, y_train)
```

```python
#obtain predictions
predicted_values = lin_model.predict(X_test)

```

```python
# import the r2 performance metric from sklearn 
from sklearn.metrics import r2_score

# Checking the R^2 value
r2_score = r2_score(y_test, predicted_values)

# Root mean square score to 2 decimal places
round(r2_score, 2)
```

#### QUESTION 13


* Normalise the dataset using the MinMaxScaler after removing the following columns: ["date", "lights"]. The target variable is "Appliances". Use a 70-30 train-test set split with a random state of 42(for reprducibility). Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions.

* What is the Mean Absolute Error in 2 d.p?



```python
#Drop of the "date" and "lights" columns
energy_data_adj = energy_data.drop(["date", "lights"], axis=1)
```

```python
#Import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

norm_energy_data = pd.DataFrame(scaler.fit_transform(energy_data_adj), columns=energy_data_adj.columns)

norm_energy_data
```

```python
#Select our Predictor and Target

y = norm_energy_data[["Appliances"]]
X = norm_energy_data.drop("Appliances", axis=1)

X
```

```python
#Split into training and testing set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=42)
```

```python
#Train a Linear reg model
linear_reg = LinearRegression()

#fit the model to the training dataset
linear_reg.fit(X_train, y_train)
```

```python
#Get predicted values

predicted_values = linear_reg.predict(X_test)
```

```python
#Import all metrics needed
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

### Mean Absolute Error

```python
mae = mean_absolute_error(y_test, predicted_values)

#Mean Absolute Error to 2 d.p
round(mae, 2)
```

### QUESTION 14

### Residual Sum of Squares

```python
rss = np.sum(np.square(y_test - predicted_values))

# Residual Sum of Squares to 2 d.p

round(rss, 2) 
```

### QUESTION 15

### Root Mean Squared Error

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))

#Root Mean Squared Error to 3 d.p
round(rmse, 3)
```

### QUESTION 16

### Coefficient of Determination or R Squared Error

```python
r2_score = r2_score(y_test, predicted_values)

#R Squared Error
round(r2_score, 2)
```

### QUESTION 17: Features Weights

```python
weights = pd.Series(linear_reg.coef_[0], index= X_train.columns).sort_values()
weights
```

RH2, RH1 are the highest and lowest respectively

### QUESTION 18: Ridge Regression model

* Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?



```python
# import ridge from sklearn

from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)
```

```python
# predicting using ridge model with test data

y_ridge_predict = ridge_reg.predict(X_test)
y_ridge_predict
```

```python
# Evaluating the ridge model through 
# RMSE

RMSE = np.sqrt(mean_squared_error(y_test,y_ridge_predict))
round(RMSE, 3)
```

The RMSE of the ridge regression model is the same as the RMSE of the linear regression model

### QUESTION 19: 

* Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
# import lasso from sklearn

from sklearn.linear_model import Lasso
```

```python
# training the lasso model

lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(X_train, y_train)
```

```python
# Obtaining the lasso model weights 

weights = pd.Series(lasso_reg.coef_, index= X_train.columns).sort_values()
weights
```

From the above, there are only 4 features with non zero:
1. RH_out  2. RH_8 
3. Windspeed 
4. RH_1
